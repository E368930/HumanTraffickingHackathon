```python
import pandas as pd
data = pd.read_csv('Documents/final_data_traff2.csv')
```

```python
data['secondarysch_teachers'] = data['secondarysch_teachers'].fillna(0)
data.head()
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=5)
from sklearn import preprocessing
#scaling the data
x= data
x = x.drop(['Unnamed: 0'], axis=1)
minmax_processed = preprocessing.MinMaxScaler().fit_transform(x)

x.head()
```

```python
df_numeric_scaled = pd.DataFrame(minmax_processed, index=x.index, columns=x.columns)
kmeans.fit(df_numeric_scaled)
kmeans.cluster_centers_.shape
#kmeans.cluster_centers_[0].shape

```

```python
features = df_numeric_scaled.columns.tolist()
print(f"Features: \n{features}")
print(len(features))
centroids = kmeans.cluster_centers_
print(f"Centroids: \n{centroids}")
print(len(centroids[0]))
```

```python
import numpy as np
sorted_centroid_features_values = np.sorted_centroid_features_values = np.take_along_axis(centroids, sorted_centroid_features_idx, axis=1)
print(f"Sorted Feature/Dimension Values for Each Centroid in Descending Order: \n{sorted_centroid_features_values}")
```
